```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
import time
```

```python
def make_file_name(path):
    result = []
    for i in range(1, 64):
        name = path+str(i)+".csv"
        result.append(name)
    return result
def get_grade_all(path, index=None):
    result = pd.DataFrame()
    file_names = make_file_name(path)
    
    for file_name in file_names:
        df = pd.read_csv(file_name, index_col=index)
        result = pd.concat([result, df])
        
    return result
def get_grade_province(file_name, index=None):
    result = pd.read_csv(file_name, index_col = index)
    return result
```

```python
path = "../data/grade/"
hg_df = pd.read_csv(path+"4.csv", index_col="SBD")
hg_df.head()
```

```python
hg_df[["KHỐI A", "KHỐI A1", "KHỐI B","KHỐI C", "KHỐI D"]].count()
```

Năm 2018, tỉnh Hà Giang có 3099 thí sinh tham dự kỳ thi TPTQG. Trong đó, có:
- 581 thí sinh thi khối A
- 574 thí sinh thi khối A1
- 570 thí sinh khối B
- 2676 thí sinh thi khối C
- 2864 thí sinh thi khối D


```python
def draw_grade_by_object(df, obj_name, min_range, max_range, step):
    plt.figure(figsize=(12, 5))
    grade_df = df[df[obj_name].notnull()]
    grade_df["class"] = np.zeros(grade_df.shape[0])
    
    for i in range(min_range, max_range, step):
        grade_df.loc[round(grade_df[obj_name]) == i,'class'] = i
        
    sns.countplot(x='class', data=grade_df)
    
    return
def draw_grade_by_subject(df, sbj_name, min_range, max_range, step):
    plt.figure(figsize=(12, 5))
    grade_df = df[df[sbj_name].notnull()]
    grade_df["class"] = np.zeros(grade_df.shape[0])
    
    for i in range(min_range, max_range, step):
        class_i = i*0.5-0.5
        grade_df.loc[((grade_df[sbj_name] <class_i) & (grade_df[sbj_name] >= class_i-0.5)),'class'] = class_i
        
    sns.countplot(x='class', data=grade_df)
    
    return
```

## Phân tích phổ điểm khối A1 của Hà Giang và cả nước

```python
draw_grade_by_object(hg_df, "KHỐI A1", 0, 30, 1)
```

Trên đây là phổ điểm khối A1 của tỉnh Hà Giang. Ta nhận thấy:

- Các thí sinh đạt trên 26 điểm lớn hơn nhiều so với lượng thí sinh đạt 24-15 điểm


```python
country_a1_df=draw_grade_by_object(country_df, "KHỐI A1", 1,30, 1)
```

- Các thí sinh đạt điểm giỏi có số lượng hợp lí

## Xét lượng thí sinh khối A1 đạt trên 27  điểm

```python
top_a1_country_df = country_df[country_df["KHỐI A1"] >=27]
top_a1_country_df.shape
```

Năm 2018, cả nước có 51 thí sinh đạt trên 27 điểm ở khối A1

```python
provinces = {'23':"Hòa Bình", '5':"Hà Giang", '1':"Hà Nội", '30':"Hà Tĩnh", '16':"Vĩnh Phúc", '2':"Sài Gòn", '19': "Bắc Ninh", '21':"Hải Dương", '7':"Lai Châu", '9':"Tuyên Quang", '10':"Lạng Sơn", '12':"Thái Nguyên",
            '15':"Phú Thọ", '28':"Thanh Hóa", '4':"Đà Nẵng", '42':"Lâm Đồng", '48':"Đồng Nai", '58':"Trà Vinh", '22':"Hưng Yên", '26':"Thái Bình", '18':"Bắc Giang", '40':"Đăk Lăk", '27':"Ninh Bình"}
```

```python
plt.figure(figsize=(20,10))
top_a1_country_df.reset_index(inplace=True)
top_a1_country_df['class'] = top_a1_country_df["SBD"].apply(lambda x:str(int(x/10**6)))
for each in provinces.keys():
    top_a1_country_df.loc[top_a1_country_df["class"] == each, 'class'] = provinces[each]
top_a1_country_df["class"].value_counts().plot(kind='pie',autopct='%1.0f%%')
plt.ylabel("")
# top_a1_country_df
```

Dựa vào biểu đồ trên, ta thấy số lượng thí sinh đạt điểm trên 27 khối A1 ở tỉnh Hà Giang và Hòa Bình chiếm tỉ trọng lớn nhất trong cả nước một cách bất thường, vượt qua cả các địa phương có truyền thống học tập tốt như Hà Nội, TP HCM, Nam Định, Nghệ An,..

## Phổ điểm môn Toán ở Hà Giang và cả nước

```python
draw_grade_by_subject(hg_df, "TOÁN", 1,22, 1)
plt.title("Phổ điểm môn Toán của tỉnh Hà Giang")
plt.show()
draw_grade_by_subject(country_df, "TOÁN",1,22, 1)
plt.title("Phổ điểm môn Toán của cả nước")
plt.show()
```

- Điểm Toán của cả nước phân bố chủ yếu trong khoảng từ 4.5 đến 6.5, từ điểm 7 trở lên ta thấy điểm càng cao thì chiều cao các cột sẽ giảm dần
- Điểm Toán ở Hà Giang tập trung chủ yếu ở khoảng điểm từ 2.5 đến 4.5, tuy nhiên tại khoảng điểm từ 8.5 đến 9.5, độ cao của các cột có sự biến đổi liên tục: Số lượng điểm 9.5 cao hơn số điểm 9.0 và số điểm 8.5 cao hơn số điểm 8.0 -> có sự bất thường tại đây vì một điều đương nhiên là ở mức điểm giỏi này, điểm càng cao thì số lượng thí sinh đạt mức điểm đó phải càng giảm mới hợp lý

## Xét lượng thí sinh đạt 9 điểm toán trở lên

```python
subject = "TOÁN"
top_math_country_df=country_df[country_df[subject]>=9]
top_math_country_df.reset_index(inplace=True)
top_math_country_df['class'] = top_math_country_df["SBD"].apply(lambda x:str(int(x/10**6)))

for each in provinces.keys():
    top_math_country_df.loc[top_math_country_df["class"] == each, 'class'] = provinces[each]
top_math_country_df.sort_values('class', ascending=False, inplace=True)
tops = top_math_country_df['class'].value_counts().keys()
for each in tops:
    if top_math_country_df.loc[top_math_country_df["class"]==each].shape[0]<8:##### MIN_ RANGE
        top_math_country_df.loc[top_math_country_df["class"]==each, 'class'] = "Other"
top_math_country_df["class"].value_counts()
plt.figure(figsize=(20,10))
# explode = (0, 0.1, 0, 0)
top_math_country_df["class"].value_counts().plot(kind='pie',autopct='%1.0f%%')
plt.ylabel("")
plt.title("BIỂU ĐỒ TỶ TRỌNG THÍ SINH ĐẠT 9 ĐIỂM TOÁN TRỞ LÊN CỦA CẢ NƯỚC NĂM 2018")
```
