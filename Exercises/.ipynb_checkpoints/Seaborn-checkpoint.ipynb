```python
def colored_star():
    size=80
    tt.color("red")
    tt.width(4)
    angle=120
    tt.begin_fill()
    for side in range(5):
        tt.forward(size)
        tt.right(angle)
        tt.forward(size)
        tt.right(72-angle)
    tt.end_fill()
```

```python
np.random.seed(123)
df = pd.DataFrame(np.random.uniform(-0.25, 0.23, size=(5,5)), 
                  columns=['var a', 'var b', 'var c', 'var d', 'var e'])
df['Dates']=pd.date_range(start=None, end=dt.datetime.today().strftime('%Y-%m-%d'),
                         periods=5).tolist()
df =df.set_index(["Dates"])
df.head()
```

```python
corr=df.corr()
corr
```

```python

# Exclude duplicate correlations by masking uper right values
mask=np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)]=True
# set background color.chart stype
sb.set_style(style='white')
f, ax = plt.subplots(figsize=(11, 9))
cmap = sb.diverging_palette(10, 250, as_cmap=True)
sb.heatmap(corr, mask=mask, cmap=cmap, square=True, linewidths=0.5, cbar_kws={'shrink':0.5}, ax=ax)
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    
    for i in range(1, 5):
        plt.plot(x, np.sin(x+i/2)*(7-i)*flip, label=i)

sb.set_style('white')
sinplot()
plt.legend()
sb.despine()
plt.show()
# sb.set_style('white')
# sb.despine()
# plt.subplots()
# sinplot()
# plt.show()
```

```python
df = sb.load_dataset('iris')
```

```python
sb.distplot(df['petal_length'], kde = True)
plt.show
```

```python
sb.distplot(df['petal_length'], hist = False, color='green')
plt.show
```

```python
sb.set_palette(sb.color_palette('Blues_r'))
sb.jointplot(x='petal_length', y='petal_width', data=df, kind = 'hex')
# most convenient way to visualize the distribution where eachobservation is
# represented in two=dimensional plot via x and y axis
```

```python
# to plot multiple pairwise bivariate distribution in dataset
sb.pairplot(df, hue='species', diag_kind='kde', kind = 'scatter', palette='husl')
```

```python
# stripplo is used when one of variable under study is categorical
sb.stripplot(x='species', y='petal_length', data=df, jitter=False)
```

```python
sb.swarmplot(x='species', y='petal_length', data=df)
# jitter can be alternated by swarmplot function
```
