```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("7.4 EXERCISE.xlsx")
data
```

```python
df =data.iloc[7:, 3:]
df.columns=["STORE OFFERS...", "Our store", "All stores"]
df=df.apply(lambda x: pd.to_numeric(x, errors ='ignore'))
df.iloc[:,1:] = df.iloc[:,1:]*100
pd.options.display.float_format = '{:.0f}%'.format
df
```

```python
df.set_index("STORE OFFERS...", inplace=True)
```
