## DATA COLLECTING

```python
from bs4 import BeautifulSoup
import requests
import json
import time
import datetime as dt
import re
import pandas as pd
import numpy as np
```

```python
url = "https://www.worldometers.info/coronavirus/#countries"
```

```python
html_text = requests.get(url).text
tree = BeautifulSoup(html_text, 'html.parser')
```

```python
date = dt.date.today()
```

```python
theads= tree.find('tr').text.replace("/\n", "/").replace("\xa0", " ").replace("\n\n", "\n").split("\n")
labels = theads[2:-3] + ["Date"]
labels
```

```python
# data = pd.DataFrame(columns = labels)
# data.to_csv("../data/coronavirus.csv", index = False)
```

```python
data = pd.read_csv("../data/coronavirus/coronavirus0520.csv")
data
```

```python
tr_all = tree.find('tbody').find_all('tr')[8:]
for tr in tr_all:
    temp = tr.text.split("\n")[2:-3]
    new_row = np.array(temp+[date])
#     print(new_row)
    data = data.append(pd.Series(new_row, index = labels), ignore_index="True")

```

```python
data.to_csv("../data/coronavirus/coronavirus0521.csv", index=False)
```
